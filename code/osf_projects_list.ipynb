```python
STARTDATE = '2022-06-26'
STOPDATE = '2023-06-26'
```

```python
# initialization

mylist = []
next_url = 'https://api.osf.io/v2/users/gswja/nodes'

while next_url:   # loop stops when next_url is None
    print(f'requesting {next_url}')
    response = reqs.get(next_url)
    response_dict = json.loads(response.text)
    next_url = response_dict['links']['next']
    print('   processing')
    
    for item in response_dict['data']:
        mydict = {}
        mydict['id'] = item['id']
        mydict['title'] = item['attributes']['title']
        mydict['description'] = item['attributes']['description']
        mydict['date_created'] = item['attributes']['date_created'][:10]
        mydict['date_modified'] = item['attributes']['date_modified'][:10]
        mylist.append(mydict)
df = pd.DataFrame(mylist)
df
```

```python
df_table = df.query(f"date_created.between('{STARTDATE}','{STOPDATE}') | date_modified.between('{STARTDATE}','{STOPDATE}')")
df_table
```
